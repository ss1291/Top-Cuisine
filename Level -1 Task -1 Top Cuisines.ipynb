# Count the number of missing values in each column

```python
missing_values_per_column = df.isnull().sum()
print(missing_values_per_column)
```

# Get unique values in the "Cuisines" column

```python
unique_cuisines = df['Cuisines'].unique()
print(unique_cuisines)
```

# Get the counts of each cuisine

```python
cuisines_num = df.Cuisines.value_counts()
cuisines_num
```

# Sort the counts of each cuisine in descending order

```python
col_sort = cuisines_num.sort_values(ascending=False)
col_sort
```

# Determine the top three most common cuisines

```python
top_cuisines = col_sort.head(3)
print("Top three most common cuisines:")
print(top_cuisines)
```

# Plotting the bar chart

```python
plt.figure(figsize=(7, 5))
top_cuisines.plot(kind='bar', color='skyblue')
plt.title('Top Three Most Common Cuisines')
plt.xlabel('Cuisines')
plt.ylabel('Counts')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

# Count the number of restaurants based on the 'Restaurant Name' column

```python
Restaurants = df['Restaurant Name'].count()
Restaurants
```

# Calculate the percentage of restaurants that serve each of the top cuisines

```python
toltal_restaurants = len(df)

cuisine_percentage = top_cuisines / Restaurants * 100

print("\nPercentage of restaurants that serve each of the top cuisines:")

print(cuisine_percentage)
```

# Plotting the pie chart

```python
colors = ['seagreen', 'goldenrod', 'crimson']
plt.figure(figsize=(5, 5))
plt.pie(cuisine_percentage, labels=cuisine_percentage.index, autopct='%1.1f%%', startangle=140, colors=colors)
plt.title('Percentage of Restaurants Serving Top Cuisines')
plt.axis('equal')
plt.show()
```
